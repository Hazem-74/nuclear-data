# __Data Scraping__

```python
import pandas as pd
import os
import requests
```

```python
# URL of the .txt file
url = "https://drhbctable.jcnp.org./Supplementary_material.txt"

# Correct the URL (remove extra dot before domain slash)
url = url.replace(".org./", ".org/")

# Send HTTP GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    with open("Data.txt", "w", encoding="utf-8") as file:
        file.write(response.text)
    print("File saved as Data.txt")
else:
    print(f"Failed to download file. Status code: {response.status_code}")
```

```python
# Load your dataset (replace 'data.csv' with your actual file)
df = pd.read_csv('Data.csv')

# Define the folder path
folder_name = "Z_data"
os.makedirs(folder_name, exist_ok=True)  # Create folder if it doesn't exist

# Ensure 'Z' column exists in the DataFrame
if 'Z' in df.columns:
    unique_Z_values = df['Z'].unique()  # Get unique values of Z

    for Z_value in unique_Z_values:
        df_Z = df[df['Z'] == Z_value]  # Filter data for the specific Z value
        filename = os.path.join(folder_name, f'Z = {Z_value}.csv')  # Create filepath
        df_Z.to_csv(filename, index=False)  # Save to CSV file
        print(f"Saved {filename}")
else:
    print("Column 'Z' not found in the dataset.")
```
